2. Create a dataset of 500 products. Each product should have a unique ID, name, description, and price.

```python
faker = Faker()

# Generate data for 500 products
products = [{'Product ID': i,
             'Name': faker.catch_phrase(),
             'Description': faker.text(),
             'Price': faker.random_number(digits=5)} for i in range(1, 501)]
```

**Gemini solution**

```python
from faker import Faker
import pandas as pd

fake = Faker()

data = []
for i in range(500):
    product_id = i + 1
    product_name = fake.catch_phrase()
    product_description = fake.paragraph(nb_sentences=3)
    product_price = round(fake.random_number(digits=2) + fake.random_digit() / 100, 2) # Generate a price with cents
    data.append([product_id, product_name, product_description, product_price])

product_catalog_df = pd.DataFrame(data, columns=['Product ID', 'Product Name', 'Product Description', 'Price'])

print("Synthetic Product Catalog:")
product_catalog_df.head()
```

3. Use Pandas to create a DataFrame from the generated data.

```python
# Create a DataFrame
product_df = pd.DataFrame(products) #pd.DataFrame is the method
print("Synthetic Product Catalog:")
product_df.head()

#Save to CSV if needed
# product_df.to_csv('synthetic_product_catalog.csv', index=False)
```
