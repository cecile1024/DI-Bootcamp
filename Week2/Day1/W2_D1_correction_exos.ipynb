**Exercise 3: Identifying Data Types**

```python
import pandas as pd
```

```python
def identify_data_types(dataset, dataset_name):
    print(f"Data Types in {dataset_name} Dataset:")
    for column in dataset.columns:
        data_type = 'Quantitative' if pd.api.types.is_numeric_dtype(dataset[column]) else 'Qualitative'
        print(f"{column}: {data_type}")
    print("\n---------------------------------------\n")
```

```python
sleep_data = pd.read_csv('/content/sleep_dataset/Time Americans Spend Sleeping.csv')
#mental_health_data = pd.read_csv('/path/to/mental_health_data.csv')
credit_card_data = pd.read_csv('/content/credit_card_dataset/clean_dataset.csv')
```

```python
identify_data_types(sleep_data, "Sleep Data")
#identify_data_types(mental_health_data, "Mental Health Data")
identify_data_types(credit_card_data, "Credit Card Approval Data")
```

**Calculate mean, median, and mode of a quantitative column (e.g., petal_length)**

```python
mean_petal_length = iris['petal_length'].mean()
median_petal_length = iris['petal_length'].median()
mode_petal_length = iris['petal_length'].mode()[0]

print(f"Mean of Petal Length: {mean_petal_length}")
print(f"Median of Petal Length: {median_petal_length}")
print(f"Mode of Petal Length: {mode_petal_length}")
```

```python
iris.describe(include='all')
```

```python
figure_histo = sns.histplot(data=iris, x="petal_length", hue="species", bins=30)
#plt.savefig('histogram.png', dpi=300)
#figure_histo.savefig('histogram.png', dpi=300)
plt.savefig("/content/drive/MyDrive/Gen-AI_2025/barplot.png", dpi=300, bbox_inches="tight")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], kde=True,bins=30) #kernel density estimation lisse l'histogramme et donne une ide de la distribution sous-jacente des donnes
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris['petal_length'], iris['petal_width'])
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

## Gold Exercise 2 : List Comprehensions With Jupyter Notebook.

```python
range_start = 1
range_end = 20
#Define the divisor (e.g., 3)

divisor = 3
#Use list comprehension to find numbers divisible by the divisor

divisible_numbers = [number for number in range(range_start, range_end + 1) if number % divisor == 0]
#Print the result

print(f"Numbers in range {range_start} to {range_end} divisible by {divisor}: {divisible_numbers}")
```

## Gold Exercise 5 : Transforming Qualitative Data Into Quantitative Data With The Titanic Dataset

```python
import pandas as pd
# Load the Titanic dataset
# Replace titanic_dataset.csv with the path to your downloaded Titanic dataset
```

```python

titanic_data = pd.read_csv('/content/train.csv')
#Check the first few rows to understand the current format of the sex column

print("Original data:")
titanic_data.head()
```

```python
#Transform the sex column
#Assuming sex is represented by 0s and 1s, where 0 = female and 1 = male
#Update this logic if the dataset uses different representations
DF= titanic_data.copy()
DF.head()
```
