### Exercise 4: Visualizing the Distribution of CGPA

```python
import seaborn as sns
import pandas as pd

# Simulated dataset
#dico
data = {'CGPA': [3.5, 3.7, 3.8, 3.9, 4.0, 3.6, 3.4, 3.8, 3.5, 3.9]}
#mettre en format pandas
df = pd.DataFrame(data)
df.head()
```

```python
#histogramme
sns.histplot(df['CGPA'], kde=True, color='green') #estmation de densité
plt.title("Distribution of CGPA")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```python
## Exercise 5: Comparing Anxiety Levels Across Different Genders
# Simulated dataset

data = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Anxiety': ['Yes', 'No', 'Yes', 'Yes', 'No']
}
#format dataFrame (la méthode .DataFrame())
df = pd.DataFrame(data)

sns.countplot(x='Gender', hue='Anxiety', data=df, palette='Set2')
plt.title("Anxiety Levels Across Genders")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
## Exercise 6: Exploring the Relationship Between Age and Panic Attacks
# Simulated dataset
data = {
    'Age': [20, 22, 24, 26, 28, 30, 32],
    'Panic Attack': [1, 0, 1, 1, 0, 0, 1]
}
df = pd.DataFrame(data)
df.head()
```

```python
sns.scatterplot( data=df, x='Age', y='Panic Attack')
plt.title("Age vs Panic Attack Occurrence")
plt.xlabel("Age")
plt.ylabel("Panic Attack (Yes=1, No=0)")
plt.show()
```
